```python
df['date']=pd.to_datetime(df['date']).dt.date
```

```python
df.head()
```

```python
df['gameday']=df['date'].apply(lambda x:x.strftime('%A'))
```

```python
df.head()
```

```python
def find_dist(df,col_name):
    date_dist=df[col_name].value_counts().reset_index()
    date_dist.columns=[col_name,"frequency"]
    sort_date_dist=date_dist.sort_values(col_name,ascending=True).set_index(col_name)
    sort_date_dist.plot(kind='bar',figsize=(20,4))
find_dist(df,'gameday')
```

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
URL = "http://www.values.com/inspirational-quotes"
r=requests.get(URL)
```

```python
soup=BeautifulSoup(r.content,'html')
print(soup.prettify())
```

```python
table=soup.find('div',attrs={'id':'all_quotes'})
```

```python
quotes=[]
```

```python
for row in table.findAll('div',attrs={'class':'col-6 col-lg-3 text-center margin-30px-bottom sm-margin-30px-top'}):
    quote={}
    quote['theme']=row.h5.text
    quote['url']=row.a['href']
    quote['img'] = row.img['src']
    quote['lines'] = row.img['alt'].split(" #")[0]
    quote['author'] = row.img['alt'].split(" #")[1]
    quotes.append(quote)
```

```python
quotes
```

```python
import csv
```

```python
with open('quotes.csv','w',newline='') as f:
    w=csv.DictWriter(f,['theme','url','img','lines','author'])
    w.writeheader()
    for q in quotes:
        w.writerow(q)
```
